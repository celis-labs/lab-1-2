Задача 2: Очистка и подготовка данных
Удалите столбцы, которые не будут использоваться в анализе.

```python
# столбцы, в которых все значения НЕ уникальны
columns_to_drop = [col for col in data.columns if data[col].nunique() == len(data)]

data = data.drop(columns=columns_to_drop)

data
```
